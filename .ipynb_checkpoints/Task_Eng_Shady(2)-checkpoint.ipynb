## Label Encoding to Categorical data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
obj_cols=[]
for i in df.columns:
    if df[i].dtype == "object":
        print("( ",i," )","Unique values are:",df[i].unique())
        print("--------------------------------------------")
        obj_cols.append(i)
print("  ")
print("Categorical Columns : ",obj_cols)
```

```python
cut_label_En = LabelEncoder()
df['cut']       = cut_label_En.fit_transform(df['cut'])

cut_ohEn = OneHotEncoder()
cut_values = cut_ohEn.fit_transform(df['cut'].values.reshape(-1,1)).toarray()

cols_cut = cut_label_En.inverse_transform(df['cut'].unique())


for i in range(len(cols_cut)):
    df[cols_cut[i]]=cut_values[:,i]
cut_values, cols_cut
```

```python
color_label_En = LabelEncoder()
df['color']       = color_label_En.fit_transform(df['color'])

color_ohEn = OneHotEncoder()
color_values = color_ohEn.fit_transform(df['color'].values.reshape(-1,1)).toarray()

cols_color = color_label_En.inverse_transform(df['color'].unique())


for i in range(len(cols_color)):
    df[cols_color[i]]=color_values[:,i]
color_values, cols_color
```

```python
clarity_label_En = LabelEncoder()

df['clarity']       = clarity_label_En.fit_transform(df['clarity'])

clarity_ohEn = OneHotEncoder()
clarity_values = clarity_ohEn.fit_transform(df['clarity'].values.reshape(-1,1)).toarray()

cols_clarity = clarity_label_En.inverse_transform(df['clarity'].unique())


for i in range(len(cols_clarity)):
    df[cols_clarity[i]]=clarity_values[:,i]
clarity_values, cols_clarity
```

```python
df.columns
```

```python
df.drop(['cut'], axis=1,inplace=True)
df.drop(['color'], axis=1,inplace=True)
df.drop(['clarity'], axis=1,inplace=True)
```

# Splitting the dataset into the Training set and Test set

```python
y = df['price'].values
dataset = df.drop(['price'], axis=1)
X = dataset.values
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
#sc_y = StandardScaler()
#y_train = sc_y.fit_transform(y_train.reshape(1, -1))
print(X_train)
print(X_test)
print(y_train)
```

```python
# Applying LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components = 1)
X_train = lda.fit_transform(X_train, y_train)
X_test = lda.transform(X_test)
```

```python
print(X_train,len(X_train))
```

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
# Visualising the Training set results
area = np.pi*6
plt.figure(figsize=(20,10))
plt.scatter(X_train, y_train, s=area, c='red', alpha=0.8)
plt.grid(True)
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('X vs price ')
plt.xlabel('LDA')
plt.ylabel('Price')
plt.show()
```

```python
# Training Polynomial Regression Model
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X_train)
poly_reg.fit(X_poly, y_train)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y_train)
print(X_poly,len(X_poly),y_train,len(y_train))
y_pred = lin_reg2.predict(X_poly)
print(y_pred,len(y_pred))
```
