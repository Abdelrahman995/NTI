```python
url = input('Enter restaurant Name url ')
html =urllib.request.urlopen(url).read()
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
soup.find_all('prw_rup')
```

```python
mydivs = soup.findAll("div", {"class": "prw_rup prw_reviews_text_summary_hsx"})
mydivs
```

```python
mydivs = soup.findAll("p", {"class": "partial_entry"})
mydivs
```

```python
reviews = [] 
for p in mydivs:
    reviews.append(str(p.text))
    #print("--------------------------------------------------------------------------------------------")
```

```python
print(len(reviews))
reviews
```

```python
import re
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
```

```python
print(reviews[0])
```

```python
for review in reviews:
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
    
```

```python
print(corpus[0])
```

```python
import csv
with open('reviews.csv', mode='w') as csvfile:
    fieldnames = ['Link', 'Review']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for review in corpus:
        writer.writerow({'Link': url, 'Review': review})
```
